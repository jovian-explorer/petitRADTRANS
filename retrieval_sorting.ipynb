```python
#importing libraries required
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/jovian-explorer/petitRADTRANS/master/JWST_WASP_96b.csv'

pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_colwidth', None)

dataset = pd.read_csv(url)
# print(dataset)
```

```python
# accessing the different columns of the csv dataset I made above
dataset.columns = ["wavelength","flux", "error"]
#storing values for each column into a separate arrays
wavelength = np.array(dataset.wavelength)            # Values of wavelength
flux = np.array(dataset.flux)                # Values of transit depth
error = np.array(dataset.error)  # Values of error

plt.scatter(wavelength, flux, s = 10,label = "Flux")

plt.legend()
plt.show()

plt.scatter(wavelength, flux, s = 10,label = "Transit depth")
plt.plot(wavelength, flux)
plt.xlabel("Wavelength")
plt.ylabel("Flux")
plt.title("Graph between Wavelength and Transit Depth")
plt.legend()
plt.show()
```
